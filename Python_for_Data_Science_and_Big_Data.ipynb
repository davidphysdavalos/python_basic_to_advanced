```python
import pandas as pd
import numpy as np
df = pd.read_csv("base_datos_2008.csv")
```

Test Chi-squared
$$\chi^2=\sum_{i=1}^k{\frac{\left(\text{observed}_i-\text{expected}_i\right)^2}{\text{expected}_i}}$$

```python
np.random.seed(0)
df=df[df["Origin"].isin(["HOU","ATL","IND"])] # Filtramos solo HOU, ATL y IND
df.sample(frac=1)
df=df[0:10000]
```

```python
df["BigDelay"]=df["ArrDelay"]>30 # Esto crea nueva columna
observados=pd.crosstab(index=df["BigDelay"],columns=df["Origin"],margins=True)
```

```python
observados #Tabla de contingencia clásica
```

```python
from scipy.stats import chi2_contingency
```

```python
test=chi2_contingency(observados)
```

```python
test
```

```python
# Estadistica, p-value y expected values
```

```python
esperados= pd.DataFrame(test[3])
```

```python
esperados
```

```python
esperados_rel=round(esperados.apply(lambda r: r/len(df)*100,axis=1),2)
```

```python
observados_rel=round(observados.apply(lambda r: r/len(df)*100,axis=1),2)
```

```python
esperados_rel
```

```python
observados_rel
```

```python
test[1]
```

__p-values__

Si $p<0.05$ significa que hay diferencias significativas: Hay relación entre las variables
Si $p>0.05$ significa que no hay diferencias significativas: No hay relación entre las variables

```python
import pandas as pd
import numpy as np
df = pd.read_csv("base_datos_2008.csv",nrows=1e5)
```

```python
x=df["ArrDelay"].dropna() # ArrDeleay y eliminamos datos faltantes
```

Cuartiles

```python
Q1=np.percentile(x,25)
Q3=np.percentile(x,75)
rangointer=Q3-Q1#Rango intercuartilico
```

Criterio clásico de selección de outliers

```python
umbralsuperior= Q3 + 1.5*rangointer
umbralinferior= Q1 - 1.5*rangointer
```

```python
umbralsuperior
```

```python
umbralinferior
```

```python
np.mean(x>umbralsuperior)
```

8% de los casos están por encima del outlier

```python
np.mean(x<umbralinferior)
```

## Lección 7: Paralelizar bucles

```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
df = pd.read_csv("base_datos_2008.csv",nrows=1e5)
```

```python
df_sub=df[["CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]]
df_sub.head(10)
```

```python
?np.isnan
```

```python
np.isnan(df["CarrierDelay"]).any() # method any returns True if any of the entries in the list/dic/array is True
```

```python
toy=df["CarrierDelay"]
np.isnan(toy)
```

```python
np.isnan(toy).any()
```

```python
max(toy)
```

```python
toy.index()
```

```python
def retraso_maximo(fila):
    if not np.isnan(fila).any():
        names=["CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]
        return names[fila.index(max(fila))]
    else:
        return "None"
```

```python
df_sub.values.tolist()[11]
```

```python
retraso_maximo(df_sub.values.tolist()[11])
```

```python
results=[]
for fila in df_sub.values.tolist():
    results.append(retraso_maximo(fila))
```

```python
results
```

Ahora paralelizamos

```python
result=Parallel(n_jobs=2,backend="multiprocessing")(map(delayed(retraso_maximo),df_sub.values.tolist()))
```

```python
result
```
